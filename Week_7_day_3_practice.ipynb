```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Reading data from an Excel file into a DataFrame
df = pd.read_excel('Capstone.xlsx', sheet_name='Sheet1')
```

```python
#Exploring the data
print(df.head())

print(df.tail())

print(df.info())

print(df.describe())

print(df.dtypes)
```

```python
print(df.dtypes)
```

```python
# graph 1 (unit price and quantity )
plt.figure(figsize=(8, 6))
plt.plot(df['unit_price'], df['quantity'], marker='o', color='blue', linestyle='-', linewidth=1) # Changed this line
plt.xlabel('unit_price')
plt.ylabel('quantity')
plt.title('Line Chart ')
plt.grid(True)
plt.show()
```

```python
# grapgh 2
plt.figure(figsize=(8, 6))
# Removed marker, linestyle, and linewidth arguments
plt.bar(df['pizza_category'], df['quantity'], width=0.5, color='blue')
plt.xlabel('pizza_category')
plt.ylabel('quantity')
plt.title('Bar Chart') # Changed title to 'Bar Chart' for better description
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# grapgh 3
plt.figure(figsize=(8, 6))
# Removed marker, linestyle, and linewidth arguments
plt.bar(df['Weekday'], df['quantity'], width=0.5, color='blue')
plt.xlabel('Weekday')
plt.ylabel('quantity')
plt.title('Bar Chart') # Changed title to 'Bar Chart' for better description
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
grouped_df = df.groupby('pizza_category').size()
print(grouped_df)
```

```python
# grapgh 4
plt.figure(figsize=(8, 6))
# Accessing 'pizza_category' from the index using .index.get_level_values()
plt.bar(grouped_df.index.get_level_values('pizza_category'), grouped_df.values, width=0.5, color='green')
plt.xlabel('pizza_category')
plt.ylabel('quantity') # Assuming you want the aggregated 'unit_price' as the y-axis
plt.title('Bar Chart')
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_df1=df.groupby(['pizza_category', 'quantity'])['unit_price'].agg(sum)
print(grouped_df1)
```

```python
# grapgh 5
plt.figure(figsize=(8, 6))
# Accessing 'pizza_category' from the index using .index.get_level_values()
plt.bar(grouped_df1.index.get_level_values('pizza_category'), grouped_df1.values, width=0.5, color='green')
plt.xlabel('pizza_category')
plt.ylabel('quantity') # Assuming you want the aggregated 'unit_price' as the y-axis
plt.title('Bar Chart')
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_df2=df.groupby(['pizza_size', 'quantity'])['unit_price'].agg(sum)
print(grouped_df2)
```

```python
# grapgh 6
plt.figure(figsize=(8, 6))
# Accessing 'pizza_category' from the index using .index.get_level_values()
plt.bar(grouped_df2.index.get_level_values('pizza_size'), grouped_df2.values, width=0.5, color='green')
plt.xlabel('pizza_size')
plt.ylabel('quantity')
plt.title('Sales. per Size')
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_df3=df.groupby(['Weekday'])['unit_price'].agg(sum)
print(grouped_df3)
```

```python
# grapgh 7
plt.figure(figsize=(8, 6))
# Accessing 'pizza_category' from the index using .index.get_level_values()
plt.bar(grouped_df3.index.get_level_values('Weekday'), grouped_df3.values, width=0.5, color='green')
plt.xlabel('Weekday')
plt.ylabel('unit_price') # Assuming you want the aggregated 'unit_price' as the y-axis
plt.title('Sales per Weekday')
plt.grid(axis="y")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
grouped_df4=df.groupby(['order_time'])['unit_price'].agg(sum)
print(grouped_df4)
```

```python
# grapgh 8
plt.figure(figsize=(8, 6))
plt.hist(grouped_df4, bins=10, color='pink', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('distribution of quantity ordered')
plt.grid(True)
plt.show()
```

```python
grouped_df5=df.groupby(['pizza_name_id'])['unit_price'].agg(sum)
print(grouped_df5)
```

```python
# grapgh 10
plt.figure(figsize=(16, 8))
plt.bar(grouped_df5.index.get_level_values('pizza_name_id'), grouped_df5.values, width=0.5, color=['green', 'blue', 'orange', 'red'])
plt.xlabel('pizza_name_id')
plt.ylabel('Total_sale')
plt.title('Total income per pizza type')
plt.grid(axis='y')
plt.xticks(rotation=90, ha='right')
plt.show()
```
